```python
#Image processing imports
import cv2
from PIL import Image, ImageFilter, ImageEnhance

#Graphs & display imports
from matplotlib import pyplot as plt , animation
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
%matplotlib inline

#Array and variable maths import
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
from scipy.io import loadmat
from scipy import signal as sig
from statistics import mean

#GUI imports
from tkinter import filedialog
import tkinter as tk

#file IO imports
import pickle
import argparse

import os

from skimage import measure
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    #gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    gray =  (1/3) * r + (1/3) * g + (1/3) * b 
    gray = gray.astype(np.uint8)
    return gray
```

```python
#thresholding :

ret,thresh2 = cv2.threshold(GrayArray,55,255,cv2.THRESH_BINARY)
detector = cv2.SimpleBlobDetector()
params = cv2.SimpleBlobDetector_Params()
```

 ## Option with skimage measure and skimage regionprops 

```python
labels = measure.label(thresh2, background=1)
propsa = measure.regionprops(labels)

fig = plt.figure(figsize = (25,35))
plt.imshow(thresh2, cmap='gray', vmin=0, vmax=1)

for label in propsa:
    object_size = len(label.coords)
    plt.plot(label.centroid[1],label.centroid[0],'o',markersize = 15, color = 'red')
    plt.text(label.centroid[1],label.centroid[0],label.area,fontsize=20,color = 'gray')
```

```python
labels = measure.label(thresh2, background=1)
propsa = measure.regionprops(labels)


fig, axs = plt.subplots(4,figsize = (25,35))
axs[0].imshow(thresh2, cmap='gray', vmin=0, vmax=1)


# detection du nombre de labels qui matchent les conditions d'aire pour plotter les images dans diffrents sous plots
foundlabel = 0

for label in propsa:
    
    object_size = len(label.coords)
    
    if label.area > 500 and label.area < 2000 :
        #Plotting markers at positions of matching pixels in blue
        axs[0].plot(label.coords[:,1],label.coords[:,0],'o',color='blue',markersize = 1)
        
        #Plotting markers at positions of centroid of labels, in red
        axs[0].plot(label.centroid[1],label.centroid[0],'o',markersize = 15, color = 'red')
        
        #Plotting text at positions of centroids of label, displaying areas in green
        axs[0].text(label.centroid[1]+10,label.centroid[0]+10,label.area,fontsize=20,color = 'green')
        
        #Printing more parameters associated with the label found
        print("Label found has major axis :{} minor axis:{} perimeter:{}".format(label.major_axis_length,label.minor_axis_length,label.perimeter))
        
        #Showing an image displaying a cropped image of the matched label, with white (True) as matched pixels, and black(False) as expluded pixels
        axs[foundlabel+1].imshow(label.filled_image,cmap='gray')
        display(label.filled_image[:,0])
        
        
        foundlabel = foundlabel +1

        #for later :
        
        #values = np.array(np.shape(label.filled_image))
        #for PX in range( np.shape(label.filled_image)[0] ) :
        #    pass

        #display(label.slice)
```
